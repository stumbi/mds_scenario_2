```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
acc = np.genfromtxt('MDSAGT_V1 2022-05-23 10-48-50/Accelerometer.csv', delimiter=',', skip_header=True)
gyr = np.genfromtxt('MDSAGT_V1 2022-05-23 10-48-50/Gyroscope.csv', delimiter=',', skip_header=True)
```

```python
print(acc)
```

```python
np.save('MDSAGT_V1 2022-05-23 10-48-50/acc_manu_1_quad', acc)
np.save('MDSAGT_V1 2022-05-23 10-48-50/gyr_manu_1_quad', gyr)
```

```python
acc = np.load('MDSAGT_V1 2022-05-23 10-48-50/acc_manu_1_quad.npy')
gyr = np.load('MDSAGT_V1 2022-05-23 10-48-50/gyr_manu_1_quad.npy')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 1, 1)
plt.plot(acc[:,1], '-', linewidth=0.75)
plt.title('Acceleration')
plt.xlabel('number of Samples')
plt.ylabel('X acceleration')

plt.subplot(3, 1, 2)
plt.plot(acc[:,2], '-', linewidth=0.75)
plt.xlabel('number of Samples')
plt.ylabel('Y acceleration')

plt.subplot(3, 1, 3)
plt.plot(acc[:,3], '-', linewidth=0.75)
plt.xlabel('number of Samples')
plt.ylabel('Z acceleration')

plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 1, 1)
plt.plot(gyr[:,1], '-', linewidth=0.75)
plt.title('Acceleration')
plt.xlabel('number of Samples')
plt.ylabel('X acceleration')

plt.subplot(3, 1, 2)
plt.plot(gyr[:,2], '-', linewidth=0.75)
plt.xlabel('number of Samples')
plt.ylabel('Y acceleration')

plt.subplot(3, 1, 3)
plt.plot(gyr[:,3], '-', linewidth=0.75)
plt.xlabel('number of Samples')
plt.ylabel('Z acceleration')

plt.show()
```
